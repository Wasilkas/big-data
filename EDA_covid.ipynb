```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

covid_data = pd.read_csv('Covid Data.csv')
covid_data.replace([97, 98, 99], np.nan, inplace=True)
covid_data.info()
```

**Легенда**  
**sex** - пол (1 - женщина, 2 - мужчина)  
**age** - возраст пациента  
**classification** - результат теста на covid. 1-3 - пациент был определен по тесту с определенной степенью заболевания, 4 и выше - пациент не являлся носителем болезни или тест не выявил заболевания  
**patient type** - место где происходил процесс выздоровления, 1 - дома, 2 - в больнице  
**pneumonia** - развилась ли у пациента пневмония, 1 - да, 2 - нет  
**pregnancy** - была ли пациентка беременна или нет, 1 - да, 2 - нет  
**diabetes** - страдает ли пациент сахарным диабетом, 1 - да, 2 - нет  
**copd** - страдает ли пациент хронической обструктивной болезнью легких, 1 - да, 2 - нет  
**asthma** - болеет ли пациент астмой, 1 - да, 2 - нет  
**inmsupr** - имеет пациент подавленный иммунитет или нет, 1 - да, 2 - нет  
**hypertension** - страдает пациент гипертонией или нет, 1 - да, 2 - нет  
**cardiovascular** - есть ли у пациента заболевания, связанные с сердцем или кровеносными сосудами, 1 - да, 2 - нет  
**renal chronic** - есть ли у пациента хронические заболевания почек, 1 - да, 2 - нет  
**other disease** - имеет ли пациент другие заболевания, 1 - да, 2 - нет  
**obesity** - страдает пациент ожирением или нет, 1 - да, 2 - нет  
**tobacco** - курит пациент или нет, 1 - да, 2 - нет  
**usmr** - обращался ли пациент в мед. учереждения, 1 - да, 2 - нет  
**medical unit** - тип учреждения национальной системы здравоохранения, которое оказывало медицинскую помощь  
**intubed** - был ли пациент подключен к аппарату ИВЛ, 1 - да, 2 - нет  
**icu** - поступал ли пациент в отделение интенсивной терапии, 1 - да, 2 - нет  
**date_died** - дата смерти(если мертв), иначе 9999-99-99
